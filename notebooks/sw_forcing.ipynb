```python
import numpy as np
import matplotlib.pyplot as plt
import tqdm
```

```python
from moviepy.editor import ImageSequenceClip
from moviepy.video.io.bindings import mplfig_to_npimage

def create_mp4(array, fps=10):
    seq = []
    for ar in tqdm.tqdm(array):
        fig, axs = plt.subplots()
        axs.imshow(ar)
        npi = mplfig_to_npimage(fig)
        plt.close()
        seq.append(npi)
    animation = ImageSequenceClip(seq, fps=fps)
    return animation.ipython_display(loop=True, autoplay=True)
```

```python
from perlin_noise import PerlinNoise

noise = PerlinNoise(octaves=5, seed=1)
xpix, ypix = 100, 100

def generate_pic(frac_t):
    return np.asarray([[noise([i/xpix, j/ypix, frac_t]) for j in range(xpix)] for i in range(ypix)])

def generate_pic_periodic(t, omega):
    return np.asarray([[noise([i/xpix, j/ypix, np.cos(omega * t),  np.sin(omega * t)]) for j in range(xpix)] for i in range(ypix)])


def generate_random(n_steps, omega=(2 * np.pi / 180)):
    pic_list = []
    for i in tqdm.trange(n_steps):
        pic_list.append(generate_pic_periodic(i, omega))
    return np.asarray(pic_list)

forcing = generate_random(180)
create_mp4(forcing)
```
